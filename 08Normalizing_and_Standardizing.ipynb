```python
import uproot
import pandas as pd
import numpy as np
import boost_histogram as bh
import matplotlib.pyplot as plt
import pickle
import atlasplots as ap
%matplotlib inline
```

```python
df_mc20_gj = pickle.load(open('picklefiles/df_mc20_gj.pickle', 'rb'))
df_mc20_jj = pickle.load(open('picklefiles/df_mc20_jj.pickle', 'rb'))
```

```python
df_gj_true = df_mc20_gj[df_mc20_gj.y_isTruthMatchedPhoton]    #only truth matched photons
df_jj_true = df_mc20_jj[~df_mc20_jj.y_isTruthMatchedPhoton]   #only NON-truth matched
```

```python
print(ap.branchlist)
print(ap.minmaxlist)
print(ap.labellist)
```

```python
#removing Rhad and Rhad1
branchlist = ap.branchlist[2:]
minmaxlist = ap.minmaxlist[2:]
labellist = ap.labellist[2:]
# np.array(df_gj_true[branchlist[1]])
```

```python
from sklearn.preprocessing import MinMaxScaler
normscaler = MinMaxScaler(feature_range=(-1, 1))

normlistgj = normscaler.fit_transform(df_gj_true[branchlist])
normlistjj = normscaler.fit_transform(df_jj_true[branchlist])


### trying with general normalization (min-max scaling)

for i in range(len(branchlist)):
    branchname = branchlist[i]
    label = labellist[i]
    minmax = minmaxlist[i]
#     normscaler.fit_transform?
#     normlistgj = normscaler.fit_transform(np.array(df_gj_true[branchname]))
#     normlistjj = normscaler.fit_transform(np.array(df_jj_true[branchname]))
    df_gj_true[branchname+'_norm'] = normlistgj[:,i]
    df_jj_true[branchname+'_norm'] = normlistjj[:,i]
    histogj = ap.makebh(df_gj_true,branchname+'_norm',(-1,1))
    histojj = ap.makebh(df_jj_true,branchname+'_norm',(-1,1))
    ap.ATLAShist(histogj,histojj,label+' norm',(-1,1))
```

```python
#### trying with min-max norm but using minmaxlist values


for i in range(len(branchlist)):
    branchname = branchlist[i]
    label = labellist[i]
    minmax = minmaxlist[i]
    normlistgj = (np.array(df_gj_true[branchname]) - minmax[0])/abs(minmax[1]-minmax[0])
    normlistjj = (np.array(df_jj_true[branchname]) - minmax[0])/abs(minmax[1]-minmax[0])
    df_gj_true[branchname+'_minmax'] = normlistgj
    df_jj_true[branchname+'_minmax'] = normlistjj
    histogj = ap.makebh(df_gj_true,branchname+'_minmax',(-0.1,1.1))
    histojj = ap.makebh(df_jj_true,branchname+'_minmax',(-0.1,1.1))
    ap.ATLAShist(histogj,histojj,label+' minmax',(-0.1,1.1))
```

```python
iskewed = [0,4,6,7]
branchskewed = [branchlist[i] for i in iskewed]
minmaxskewed = [minmaxlist[i] for i in iskewed]
labelsskewed = [labellist[i] for i in iskewed]
```

```python
#### making the skewed variable distributions more uniform:

for i in range(len(branchskewed)):
    branchname = branchskewed[i]
    label = labelsskewed[i]
    minmax = minmaxskewed[i]
    skewedlistgj = np.sqrt(np.array(df_gj_true[branchname+'_minmax']))
    skewedlistjj = np.sqrt(np.array(df_jj_true[branchname+'_minmax']))
    df_gj_true[branchname+'_minmax_unskew'] = skewedlistgj
    df_jj_true[branchname+'_minmax_unskew'] = skewedlistjj
    histogj = ap.makebh(df_gj_true,branchname+'_minmax_unskew',(0,1.1))
    histojj = ap.makebh(df_jj_true,branchname+'_minmax_unskew',(0,1.1))
    ap.ATLAShist(histogj,histojj,'sqrt{'+label+' minmax',(-0.1,1.1))
```
